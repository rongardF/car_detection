```python
# COPY
new_mat = scipy.io.loadmat(f"{devkit_path}/cars_train_annos.mat")
trial = np.hstack(new_mat['annotations'])
trial_df = pd.DataFrame(trial)
trial_df['file_name'] = trial_df['fname']
trial_df['file_name'] = trial_df['file_name'].map(np.array2string)
trial_df['file_name'] = trial_df['file_name'].map(lambda x:x.lstrip("['").rstrip("']"))

trial_df
# min(trial_df["class"])
```

```python
orig_mat = scipy.io.loadmat(original_annos_path)
# orig_annos
orig_annos = np.hstack(orig_mat['annotations'])
orig_df = pd.DataFrame(orig_annos)
orig_df['class_number'] = orig_df['class'].astype(float)

names= orig_mat['class_names'].copy()
names_df = pd.DataFrame(names).T
names_df['class_number'] =names_df.index+1
names_df['car_name'] = names_df[0].map(lambda x:np.array2string(x).lstrip("['").rstrip("']"))
names_df = names_df.drop(columns = 0)

merged_df= orig_df.merge(names_df, left_on='class_number', right_on = 'class_number')

# merged_df
merged_df.loc[merged_df["class_number"]==1.0]
```
